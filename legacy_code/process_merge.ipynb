```python
import pandas as pd
import numpy as np
import re

# Reload the original data
data = pd.read_csv('data/final_raw_data.csv')

# Function to check if the event_name is of the form "경험치 n배 쿠폰 m장 지급"
def experience_coupon_event(event_name):
    return bool(re.match(r'익스트림 성장의 비약 \d+개 지급', event_name))


# Apply the function to the 'event_name' column
data['is_experience_coupon'] = data['event_name'].apply(experience_coupon_event)

date_set = set()

temp_df = data[data['is_experience_coupon']]

for i in range(temp_df.shape[0]):
    for j in range(temp_df.shape[1]):
        if pd.isna(temp_df.iloc[i, j]):
            break
        if j != 0:
            date_set.add(temp_df.iloc[i, j])

date_list = list(sorted(date_set))
index_list = list(map(str, range(len(date_list))))

new_row = pd.Series(['익스트림 성장의 비약 지급'] + date_list, index = ['event_name'] + index_list)

new_row
```

```python
data = data[data['is_experience_coupon'] == 0]
data = data.append(new_row, ignore_index = True)
data
```

```python
data = data.iloc[:, :-1]
data
```

```python
data.to_csv('data/final_raw_data.csv', index=False)
```
