```python
import pandas as pd

# Load the csv file
df = pd.read_csv('data/sundaymaple_date.csv', index_col= 0)

def slice_if_not_nan(val):
    if pd.isna(val):
        return val
    else:
        return val[:4]

df.applymap(slice_if_not_nan)
```

```python


df_melted = df.melt(id_vars='event_name', value_name='date')
df_melted.dropna(inplace=True)
```
