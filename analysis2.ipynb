```python
import pandas as pd
import numpy as np
from datetime import date
import datetime

df = pd.read_csv('data/final_date.csv')

class EventAnalyzer:
    def __init__(self, df):
        self.df = df
#------------------------------------------------------------
# private method
#------------------------------------------------------------

    def _get_combined_dates(self, event_names):
        #여러 이벤트 이름에 해당하는 날짜들을 하나의 리스트로 합쳐 반환합니다.
        combined_dates = []
        for event_name in event_names:
            row = self.df[self.df['event_name'] == event_name].iloc[0, 1:].dropna()
            dates = [datetime.datetime.strptime(date, '%Y-%m-%d').date() for date in row]
            combined_dates.extend(dates)
        
        combined_dates.sort()
        return combined_dates
    
#------------------------------------------------------------
# public method
#------------------------------------------------------------

    def event_intervals(self, event_names):
        #합쳐진 여러 이벤트 이름에 대해 연속된 날짜 간의 간격을 반환합니다.
        dates = self._get_combined_dates(event_names)
        intervals = [(dates[i+1] - dates[i]).days for i in range(len(dates)-1)]
        return intervals

    def event_counts_by_year(self, event_names):
        #합쳐진 여러 이벤트 이름에 대해 연도별 발생 횟수를 반환합니다.
        dates = self._get_combined_dates(event_names)
        counts = {}
        for date in dates:
            year = date.year
            if year in counts:
                counts[year] += 1
            else:
                counts[year] = 1
        return counts
    
    def get_common_dates(self, event_name1, event_name2):
        #두 이벤트 이름에 대해 공통된 날짜를 리스트로 반환합니다.
        dates1 = set(self.df[self.df['event_name'] == event_name1].iloc[0, 1:].dropna())
        dates2 = set(self.df[self.df['event_name'] == event_name2].iloc[0, 1:].dropna())
        return sorted(list(dates1.intersection(dates2)))
    
    def row_data_count(self):
        # 각 row의 데이터 개수를 반환합니다 (NaN이 아닌 값의 개수).
        counts = {}
        for index, row in self.df.iterrows():
            event_name = row['event_name']
            count = row[1:].notna().sum()  # NaN이 아닌 값의 개수를 계산
            counts[event_name] = count
        return counts
    

# 테스트
analyzer = EventAnalyzer(df)
common_dates_result = analyzer.get_common_dates('스타포스 5, 10, 15성에서 강화 시도 시 성공확률 100%', '스타포스 강화 비용 30% 할인')
common_dates_result
```
