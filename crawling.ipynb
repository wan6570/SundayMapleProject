```python
import requests
from bs4 import BeautifulSoup

# 웹페이지에서 데이터를 크롤링합니다.
url = "https://www.inven.co.kr/board/maple/2304/20563"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# 데이터를 저장할 딕셔너리를 생성합니다.
data = {}

# "가나다순"이라는 텍스트가 포함된 <font size="2"> 태그를 찾습니다.
font_tag = soup.find('font', {'size': '2'}, string="(가나다순)")

# font 태그 아래의 모든 텍스트를 순차적으로 추출합니다.
strings = list(font_tag.find_all_next(string=True))

# 이벤트 이름과 날짜를 추출합니다.
event = None
dates = []
for string in strings:
    string = string.strip()
    if string == '날짜별':
        break
    if string == '(단종)':
        continue
    if string.startswith('-'):
        if event is not None:
            data[event] = dates
        event = string
        dates = []
    elif string and event is not None:
        dates.append(string)

# 마지막 이벤트를 데이터에 추가합니다.
if event is not None:
    data[event] = dates

print(data)
```

```python
import csv

# 데이터를 CSV 파일로 저장합니다.
with open('data.csv', 'w', newline='', encoding='utf-8-sig') as f:
    writer = csv.writer(f)
    for event, dates in data.items():
        writer.writerow([event] + dates)
```
